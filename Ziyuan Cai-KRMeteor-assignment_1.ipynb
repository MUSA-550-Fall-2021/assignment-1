```python
import pandas as pd
```

```python
ZHVI = pd.read_csv("E:/graduate/musa550/assignment/ass1/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
```

```python
want_state = ["PA"]
selection_state =  ZHVI['State'].isin(want_state)
ZHVI_trimmed_state = ZHVI.loc[selection_state]
```

```python
want_city = ["Philadelphia"]
selection_city =  ZHVI_trimmed_state['City'].isin(want_city)
ZHVI_trimmed = ZHVI_trimmed_state.loc[selection_city]
```

```python
columns = ['RegionID', 'SizeRank', "RegionType", "StateName", "State", "Metro", "CountyName", "City"]
ZHVI = ZHVI_trimmed.drop(columns, axis=1)
```

```python
ZHVI_final = pd.melt(
    ZHVI, 
    id_vars=["RegionName"], 
    var_name="Date",
    value_name="ZHVI"
)
```

```python
ZHVI_final["Date"] = pd.to_datetime(ZHVI_final["Date"])
date_strings = ZHVI_final["Date"].dt.strftime("%Y")
ZHVI_final["Date"] = date_strings
```
